Функции

```python
import requests
from bs4 import BeautifulSoup

# Функция для парсинга описания из url страницы аниме
def description_from_url(url): 
    print(url)
    headers = {
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.9; rv:45.0) Gecko/20100101 Firefox/45.0'
      }
    try:
        rs = requests.get(url, headers = headers)
        soup = BeautifulSoup(rs.content, 'html.parser')
        div_description = soup.select_one('div.b-text_with_paragraphs')
#     print(rs)
    # print(soup)
#     print(div_description)
        soup_description = BeautifulSoup(div_description.text)
        description = soup_description.select_one('p').text
#     print(description)
    except Exception:
        return None
    
    return description
    
```

```python
import csv
import time

# Функция перебора и сохранения данных об анме в файл
def save_anime(api, start_page=1, end_page=2, file_name="data.csv"):
    # Открываем файл
    f = open(file_name, mode="w", encoding='utf-8')
    file_writer = csv.writer(f, delimiter = ",", lineterminator="\r")
    file_writer.writerow(["Название", "Описание", "Оценка"])
        
    # Проходимся по аниме
    for i in range(start_page, end_page):
        # формируем выборку со страницы
        itr = 1
        while itr <= 3:
            try:
                anime_list = api.animes.GET(order='popularity', status = 'released', limit=50, censored='false', page=i)
                itr = 4
            except Exception as e:
                anime_list = None
                print(f"Ошибка считывания с {i} страницы: {e}")
                # пробуем повторить через 120 секунд
                time.sleep(120)
                itr += 1
            
        # завершаем перебор, если аниме больше нет или возникла ошибка
        if not anime_list:
            break

        # формируем выборку со страницы
        for anime in anime_list:
            url = 'https://shikimori.one/' + anime['url']

            name = anime['name']
            description = description_from_url(url)
            score = anime['score']
            if description is None:
                continue
                
            # сохраняем
            file_writer.writerow([name, description, score])
            
    f.close()
    
```
